# **CONSULTAS SQL (A | B | M | L)** 

Las consultas SQL son los "dialogos" o "preguntas" que se generan entre el usuario y el sistema gestor de bases de datos donde se encuentran almacenados ciertos datos

# **De Lectura**

- **SELECT**: Sentencia utulizada para especificar a que atributo (dato) se pretende acceder

- **FROM**: se utiliza con SELECT, para especificar desde que tabla (entidad) se pretende traer el dato

- **WHERE**: Sentencia para proponer la condicion especifica que debe cumplir el dato que se pretende traer (no es una sentencia obligatoria)

# **De Escritura** (Alta)

- **INSERT INTO**: Es utulizada para especificar en que tabla se pretende insertar un dato

- **VALUES**: Se utiliza en conjunto con INSERT INTO para especificar que valores iran en la tabla
Los valores siempre van entre parentesis separados por coma (,) y en caso de que sea string se pone entre comillas

# **De Modificacion**

- **UPDATE**: Es utilizada para especificar en que tabla se pretende modificar un dato

- **SET**: Se utiliza en conjunto con UPDATE para especificar cual sera el nuevo valor/dato

- **WHERE**: Sentencia para proponer la condicion especifica que debe cumplir el dato de referencia del registro que se pretende modificar (No es una sentencia obligatoria, pero si recomendada)

# **De Baja**

- **DELETE FROM**: Es utilizada para modificar un registro de una tabla de forma permanente 

- **WHERE**: Sentencia para proponer la condicion especifica que debe cumplir el dato de referencia del registro que se pretende dar de baja (No es una sentencia obligatoria pero si recomendada)





# **ORDER y GROUP BY** (De Orden y/o Agrupamiento)

**SE USAN EN CONJUNTO CON EL SELECT**

- **ORDER BY**: Es utilizada para especificar por que criterio se pretende ordenar los registros de una tabla

Con ORDER BY tambien se usa ASC Y DESC para ordenar de forma ascendente o descendente

            SELECT * 
            FROM empleado
            ORDER BY apellido ASC

En esta consulta, traemos todos los datos de la tabla empleados, y ordenamos teniendo en cuenta el apellido en orden ascendente (a->z)

- **GROUP BY**: Es utilizada para especificar por que criterio se puede agrupar los registros de una tabla

            SELECT cargo
            FROM empleado
            GROUP BY cargo

En esta consulta si hay varios empleados con el mismo puesto/cargo, solo se mostrara un cargo respectivo

# **LEFT JOIN**
Muestra el conjunto completo de los registros de la tabla A, con los registros que coincidan de la tabla B (si es que los hay)



**Ej:**

            SELECT *

            LEFT JOIN departamento

            ON empleado.id_departamento = departamento.id_departamento

Traeme todo de empleado, uniendo por izquierda con departamento donde coincidan las id departamento

# **RIGHT JOIN**
Es el reflejo del Left Join. Muestra el conjunto completo de los registros de la tabla B, con los registros que coincidan de la tabla A (si es que los hay)



**Ej:**

            SELECT *
